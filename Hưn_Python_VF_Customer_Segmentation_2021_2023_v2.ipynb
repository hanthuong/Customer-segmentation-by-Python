# 3. Data cleaning and preprocessing

```python
# drop duplicates
```

```python
data = data.drop_duplicates()
data.info()
```

```python
# drop column: Index
```

```python
data = data.drop(columns=['Index'])
data.info()
```

```python
# remove outliers: Quantity by #Capping
```

```python
q3 = data['Quantity'].quantile(0.75)
q1 = data['Quantity'].quantile(0.25)

IQR = q3 - q1

quantity_lower_limit = q1 - 1.5*IQR
quantity_upper_limit = q3 + 1.5*IQR

quantity_lower_limit, quantity_upper_limit
```

```python
data['Quantity_RO'] = pd.DataFrame(np.where(data['Quantity']>quantity_upper_limit, quantity_upper_limit,
                      np.where(data['Quantity']<quantity_lower_limit, quantity_lower_limit, data['Quantity'])), columns=['Quantity_RO'])
data.head()
```

```python
sns.boxplot(data=data[['Quantity','Quantity_RO']])
```

# 5. Feature creation

```python
# RFM model
```

```python
# Monetary = InvoiceValue
```

```python
# Recency = last_InvoiceDate - IssuedDate
```

```python
last_InvoiceDate = data.groupby('CustomerCode',as_index=False)['IssuedDate'].max()
last_InvoiceDate.rename(columns={'IssuedDate': 'last_InvoiceDate'},inplace=True)
last_InvoiceDate.head()
```

```python
data = data.merge(last_InvoiceDate,on='CustomerCode')
data['Recency'] = ((data['last_InvoiceDate'] - data['IssuedDate']) / np.timedelta64(1,'M')).astype('int64')
data.head()
```

```python
np.sort(data['Recency'].unique())
```

```python
# Frequency = Invoice counts
```

```python
df = data.groupby(['CustomerCode']).agg({'Recency':'max','Invoice':'nunique','InvoiceValue':'sum'}).reset_index()
df.rename(columns={'Invoice':'Frequency','InvoiceValue':'Monetary'},inplace=True)
df.head()
```

```python
df = df.drop(df[df['CustomerCode'] == 'ANONYMOUS'].index)
df.head()
```

```python
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
```

```python
sns.pairplot(data=df[['Recency','Frequency','Monetary']])
```

```python
X = df[['Frequency','Monetary']]
X.head()
```

```python
kmean = KMeans(n_clusters=3)
kmean.fit(X)
y_pred = kmean.predict(X)
y_pred
```

```python
kmean.inertia_
```

```python
sse = []
for i in range(2,30):
  kmean = KMeans(n_clusters=i)
  kmean.fit(X)
  sse.append((i, kmean.inertia_))
sse
```

```python
sse = np.array(sse)
plt.plot(sse[:,0], sse[:,1])
plt.show()
```

```python
# clusters = 10
```

```python
kmeans = KMeans(n_clusters=10)
kmeans.fit(X)
y_kmean = kmeans.predict(X)
y_kmean
```

```python
df_new = pd.concat([X, pd.Series(y_kmean)],axis=1)
df_new = df_new.rename(columns={0:'y_kmeans'})
df_new.head()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data=X,
                x='Frequency',
                y='Monetary',
                hue=y_kmean,
                s=100,
                alpha=0.7)
plt.show()
```
